```python
import csv
import sys
import os
import dateutil
from datetime import *
import pandas
from dateutil.rrule import *


##################################################
##################################################

day = 0
def first_ride(reader, day=1):
   for row in reader:
       starttime = dateutil.parser.parse(row['starttime'])
       if starttime.day == day:
          day += 1
          yield row

##################################################
##################################################

if __name__=='__main__':
    if len(sys.argv)<2:
        sys.stderr.write('USAGE: python %s <INPUT_CSV>\n' % sys.argv[0])
        sys.exit(1)
    
    with open(os.sys.argv[1], 'r') as fi:
        reader = csv.DictReader(fi)
        for row in first_ride(reader):
            print ','.join(map(row.get, reader.fieldnames))
            
```
